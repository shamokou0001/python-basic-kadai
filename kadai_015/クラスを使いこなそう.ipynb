```python
# Humanクラスの定義
class Human:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def printinfo(self):
        print("名前:", self.name)
        print("年齢:", self.age)


# Humanクラスのインスタンスを変数に代入
human1 = Human("太郎", 20)

# メソッドを使用して名前と年齢を出力
human1.printinfo()
```
